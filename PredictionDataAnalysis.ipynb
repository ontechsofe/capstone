```python
from matplotlib import pyplot as plt
from pandas import DataFrame, read_csv
import numpy as np
import math
%matplotlib inline
```

```python
def build_plot_xy(data_path: str, testing: str, rolling_average: bool = False, predict_above: bool = True):
    df = read_csv(data_path)
    predicted_correctly = len([x for x in df[testing] if x > 0.5]) 
    if not predict_above: 
        predicted_correctly = len([x for x in df[testing] if x <= 0.5])
    total_data_points = df.shape[0]
    x = list(range(total_data_points))
    y = df[testing]
    if rolling_average:
        y = df[testing].rolling(window=20).mean()
    print(f"PREDICTED CORRECTLY: {predicted_correctly}")
    print(f"TOTAL DATA POINTS: {total_data_points}")
    print(f"ACCURACY: {(predicted_correctly/total_data_points) * 100:.2f}%")
    return x, y
```

# Focussed

```python
x, y = build_plot_xy(
    data_path="data/predictions/JOSH-FOCUS.csv",
    testing="focus_prediction",
)
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.scatter(x, y)
ax.set_title("Focus Prediction - Tester 00")
ax.set_xlabel("Data Point")
ax.set_ylabel("Focus Prediction")
```
